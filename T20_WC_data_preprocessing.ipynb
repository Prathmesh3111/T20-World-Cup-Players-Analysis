<h4 style="color:blue">(2) Process Batting Summary</h4>

```python
df2 = pd.read_json('t20_json_files/t20_wc_batting_summary.json')
```

```python
all_matches = []

for match in df2.values:
    all_matches.extend(match[0])
```

```python
df2 = pd.DataFrame(all_matches)
df2.head(12)
```

```python
df2['out/not_out']  = np.where(df2['dismissal'] == "", 'not_out', 'out')
```

```python
df2.head(30)
```

```python
df2.drop('dismissal', axis=1, inplace=True)
```

```python
df2.head()
```

```python
def clean_name(name):
    clean_name = ''
    for i in name:
        if i.isalpha() or i==" ":
            clean_name += i
        else:
            break
            
    return clean_name
```

```python
df2["batsmanName"] = df2["batsmanName"].apply(clean_name)
```

```python
df2.head()
```

```python
df1.head()
```

```python
match_id_dict = {}

for index, row in df1.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
    
match_id_dict
```

```python
df2['match_id'] = df2['match'].map(match_id_dict)
```

```python
df2.sample(10)
```

```python
df2.to_csv('t20_csv_files/batting_summary.csv', index=False)
```

```python
df3 = pd.read_json('t20_json_files/t20_wc_bowling_summary.json')
```

```python
matches2 = []

for roc in df3['bowlingSummary'].values:
    matches2.extend(roc)
```

```python
df3 = pd.DataFrame(matches2)
```

```python
df3.head()
```

```python
df3["bowlerName"] = df3["bowlerName"].apply(clean_name)
```

```python
df3['match_id'] = df3['match'].map(match_id_dict)
```

```python
df3.to_csv('t20_csv_files/bowling_summary.csv', index=False)
```
